```python
nlp = spacy.load("en_core_web_lg")
```

```python
data.head(3)
```

```python
X = data['description']
Y = data['product_category']
```

```python
def lemmatize(text):
    doc = nlp(text)
    lemma_text = [token.lemma_.lower() for token in doc]
    return " ".join(lemma_text)
    
```

```python
x_data = X.apply(lemmatize).astype("string")
```
