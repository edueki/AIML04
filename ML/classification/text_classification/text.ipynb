```python
nlp = spacy.load("en_core_web_lg")
```

```python
X = data['description']
Y = data['product_category']
```

```python
def lemmatize(text):
    doc = nlp(text)
    lemma_text = [token.lemma_.lower() for token in doc]
    return " ".join(lemma_text)
    
```

```python
x_data = X.apply(lemmatize).astype("string")
```

```python
x_data
```

```python
STOP_WORDS  = nlp.Defaults.stop_words
print (STOP_WORDS)
```

```python
SAFE_WORDS = {"no", "not", "nor"}
SAFE_STOPWORDS = set(STOP_WORDS) - SAFE_WORDS
def remove_stopwords_spacy(text: str) -> str:
    doc = nlp(str(text))
    tokens = [t.text for t in doc if t.is_alpha and t.text.lower() not in SAFE_STOPWORDS]  # <-- per-token check
    return " ".join(tokens)
```

```python
x_data_new = x_data.apply(remove_stopwords_spacy)
```

```python
x_data_new
```

```python
import numpy as np
def doc_vector_safe(text: str):
    doc = nlp(str(text))                 # nlp(x)
    if doc.vector_norm:                  # doc has a non-zero vector
        return doc.vector

X_vec = np.vstack(x_data_new.apply(doc_vector_safe).to_numpy())
print("Embeddings shape:", X_vec.shape)  # -> (n_samples, 300) for _md
```

```python
X_vec
```

```python
Y
```

```python
Y_new = Y.map({'Electronics':0,'Clothing':1})
```

```python
Y_new
```

```python
!python3 -m pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
Xtr, Xte, ytr, yte = train_test_split(X_vec, Y_new, test_size=0.2, random_state=42)
clf = LogisticRegression(max_iter=2000).fit(Xtr, ytr)
print("Accuracy:", clf.score(Xte, yte))
```

```python
predictions = clf.predict(Xte)
```
