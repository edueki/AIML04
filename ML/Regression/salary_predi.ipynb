```python
from matplotlib import pyplot as plt
plt.figure(figsize=(6,6))
plt.scatter(x,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience vs. salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model.coef_
```

```python
model.intercept_
```

```python
emp_data = pd.read_csv("employee_exp.csv")
```
